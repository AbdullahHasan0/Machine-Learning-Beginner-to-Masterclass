```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("canada_per_capita_income.csv")
df
```

```python
%matplotlib inline
plt.xlabel='price'
plt.ylabel='per_capita_income'
plt.scatter(df.year,df.per_capita_income,color="purple")
```

```python
new_df=df.drop("per_capita_income",axis='columns')
new_df
```

```python
model = linear_model.LinearRegression()
```

```python
model.fit(new_df,df.per_capita_income)
```

```python
model.predict([[2023]])
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_pred = model.predict(new_df)
```

```python
plt.scatter(new_df, df.per_capita_income, label='Data')
plt.plot(new_df, y_pred, color='red', linewidth=2, label='Best fit line')
plt.legend()
plt.show()
```
