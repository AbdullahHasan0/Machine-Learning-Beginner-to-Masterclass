```python
import pandas as pd
from sklearn import tree
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('titanic.csv')
data
```

```python
data=data.drop('PassengerId',axis = 'columns')
data=data.drop('Name',axis = 'columns')
data=data.drop('SibSp',axis = 'columns')
data=data.drop('Parch',axis = 'columns')
data=data.drop('Ticket',axis = 'columns')
data=data.drop('Cabin',axis = 'columns')
data=data.drop('Embarked',axis = 'columns')
data

```

```python
target = data['Survived']
target
data
```

```python
inputs = data.drop('Survived',axis = 'columns')
inputs
```

```python
le_Sex = LabelEncoder()
inputs['Sex_n']=le_Sex.fit_transform(inputs['Sex'])
inputs
```

```python
inputs_final=inputs.drop('Sex',axis = 'columns')
inputs_final
```

```python
median = inputs_final.Age.mean()
inputs_final.Age=inputs_final.Age.fillna(median)
```

```python
X_train,X_test,y_train,y_test=train_test_split(inputs_final,target,test_size=0.3)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
model = tree.DecisionTreeClassifier()
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
